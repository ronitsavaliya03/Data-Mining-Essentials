# Step 1: Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('datasets/OnlineRetail.csv', encoding='ISO-8859-1')
df.head(5)
```

```python
df
```

# Step 2: Drop Missing Invoices or Descriptions

```python
df = df.dropna(subset=['InvoiceNo', 'Description'])
df
```

# Step 3: Remove Canceled Invoices <BR> Start with letter 'c'

```python
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
df = df[~df['InvoiceNo'].str.startswith('C')]
df
```

# Step 4: Keep Only Positive Quantities

```python
df = df[df['Quantity'] > 0]
df
```

# Step 5: Standardize Item Descriptions | Lower case and remove space

```python
df['Description'] = df['Description'].str.lower().str.strip()
df
```

# Step 6: Filter for a Single Country (e.g., United Kingdom)

```python
df = df[df['Country'] == "United Kingdom"]
df
```

# Step 7: Remove Duplicates

```python
# not done
d = df.drop_duplicates()
d
```

# Step 8: Add TotalPrice column , TotalPrice = Quantity *  UnitPrice

```python
df["TotalPrice"]=df["Quantity"]*df["UnitPrice"]
df
```

# Step 9 : Check for Top 10 Selling Products

```python
df.groupby('Description')['TotalPrice'].sum().sort_values(ascending=False).head(10)
```

# Step 10 : Identify Most Active Customers

```python
df.groupby('CustomerID')['InvoiceNo'].count().sort_values(ascending=False).head(10)
```

# Step 11: Time-Based Analysis | Convert InvoiceDate to datetime and check popular months/hours.

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

df['Month'] = df['InvoiceDate'].dt.month
df['Hour'] = df['InvoiceDate'].dt.hour

popular_months = df['Month'].value_counts().sort_index()
popular_hours = df['Hour'].value_counts().sort_index()

monthly_sales = df.groupby('Month')['TotalPrice'].sum()

# print(popular_months)
# print(popular_hours)
# print(monthly_sales)

plt.bar(monthly_sales.index, monthly_sales, color='blue')
plt.title("Monthly Sales Analysis")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1, 13))  
plt.tight_layout()
plt.show()
```

# Step 12: Convert to Basket Format

```python
basket = (df
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
          .unstack()
          .fillna(0))

basket = basket.applymap(lambda x: 1 if x > 0 else 0)
```

# Step:13 : Save Basket Format into CSV file

```python
basket.to_csv("basket_format.csv")
print("Basket format saved to 'basket_format.csv'")
```
