```python
df1=df.groupby("Attrition_Flag")["Credit_Limit"].mean()
df1
```

```python
df2=df.groupby(["Attrition_Flag", "Gender"])["Total_Trans_Amt"].mean()
df2
```

```python
df3=df.groupby("Income_Category")["CLIENTNUM"].count()
df3
```

```python
df4=df.groupby("Education_Level").agg({
    "Credit_Limit": "mean",
    "Total_Trans_Amt": "sum"
})
df4
```

```python
attrition_counts = df.groupby(["Marital_Status", "Attrition_Flag"])["CLIENTNUM"].count().unstack()
attrition_percentage = attrition_counts.div(attrition_counts.sum(axis=1), axis=0) * 100
attrition_percentage
```

```python
df5=df.groupby("Months_on_book")["Total_Trans_Amt"].mean()
df5
```

```python
df7=df[df["Customer_Age"] > 50].groupby("Card_Category")["Avg_Utilization_Ratio"].mean()
df7
```

```python
def iqr(x):
    return x.quantile(0.75) - x.quantile(0.25)

df8=df.groupby("Attrition_Flag")["Total_Trans_Amt"].agg(["median", iqr])
df8
```
